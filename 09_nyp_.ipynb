# Kalıtım



Kalıtım (inheritance), nesne yönelimli programlamada bir sınıfın başka bir sınıftan özelliklerini (metodlar ve değişkenler) miras almasını sağlayan bir mekanizmadır. Bu, kod tekrarını azaltmaya ve kod organizasyonunu iyileştirmeye yardımcı olur.

Python'da kalıtım, bir sınıf tanımlarken parantez içinde başka bir sınıfın adını belirterek kullanılır. Bu, belirtilen sınıfın tüm özelliklerini yeni sınıfa aktarır. Yeni sınıf, 'alt sınıf' veya 'türetilmiş sınıf' olarak adlandırılırken, özellikleri aktarılan sınıf 'üst sınıf' veya 'temel sınıf' olarak adlandırılır.


```python
class EvcilHayvan():
    def __init__(self, ad, agirlik) -> None:
        self.ad = ad
        self.agirlik = agirlik
        
    def yemek_ye(self):
        print("Hayvanınız yemek yedi")
    def ses_cikar(self):
        print("Hayvanınız ses çıkardı") # Tüm hayvanlar aynı sesi mi çıkartır?
        
class Kus(EvcilHayvan): # Kuş sınıfı EvcilHayvan sınıfından türemiştir.
    tur = "Kuş"
    
class Kopek(EvcilHayvan): # Kuş sınıfı EvcilHayvan sınıfından türemiştir.
    tur = "Köpek"
    
class Kedi(EvcilHayvan): # Kuş sınıfı EvcilHayvan sınıfından türemiştir.
    tur = "Kedi"


kus_1 = Kus("Twety", 152)
kedi_1 = Kedi("Tom", 3200)
kopek_1 = Kopek("Spike", 7800)

for hayvan in (kus_1, kopek_1, kedi_1):
    print("*******************")
    print("Tür: ", hayvan.tur)
    print("Ad: ", hayvan.ad)
    print("Ses çıkar")
    hayvan.ses_cikar()
```



Aşağıdaki örnekte iki sınıf tanımlanmıştır: Calisan ve Yonetici. Yonetici sınıfı, Calisan sınıfından miras alır, bu da Yoneticinin Calisanın tüm özelliklerine ve metodlarına sahip olduğu anlamına gelir.

Calisan Sınıfı:

- *--init--* metodu, bir Calisan nesnesi oluşturulurken çağrılır ve nesnenin başlangıç durumunu ayarlar. Bu metod, çalışanın adı, maaşı, departmanı ve ID'si gibi bilgileri parametre olarak alır ve bu bilgileri nesne değişkenlerine (*self.ad*, *self.maas*, *self.departman*, *self.id*) atar.

- *ise_git()* metodu, çalışanın işe gittiğini simgeleyen bir mesaj yazdırır.

- *tatil()* metodu, çalışanın tatilde olduğunu simgeleyen bir mesaj yazdırır.

Yonetici Sınıfı:

- Yonetici, Calisan sınıfından türetilmiştir ve Calisan sınıfının tüm özelliklerini ve metodlarını miras alır.
- *--init--* metodu, bir Yonetici nesnesi oluşturulurken çağrılır. Bu metod, Calisan'ın *--init--* metodunu super().__init__(ad, maas, departman, id) ifadesiyle çağırır ve Yonetici nesnesi için de ad, maaş, departman ve ID bilgilerini ayarlar. 
- Yöneticinin sorumlu olduğu ekibi belirten bir ekip parametresi alır ve bu bilgiyi *self.ekip* nesne değişkenine atar.

Yonetici sınıfı, Calisan sınıfından özellikle farklılaştırılmış veya üzerine yazılmış herhangi bir metoda sahip değildir, ancak gerektiğinde Calisan sınıfındaki metodları kendi ihtiyaçlarına göre özelleştirebilir.

```python
class Calisan():
    def __init__(self, ad, maas, departman, id) -> None:
        self.ad = ad
        self.maas = maas
        self.departman = departman
        self.id = id
        
    def ise_git(self):
        print("Bugün de işe geldik.")
        
    def tatil(self):
        print("Bugün yatıştayız.")
        
class Yonetici(Calisan):
     def __init__(self, ad, maas, departman, id, ekip) -> None:
        super().__init__(ad, maas, departman, id)
        self.ekip = ekip
```

## super() fonksiyonu

super() fonksiyonu, Python'da bir sınıfın üst sınıfına (veya ebeveyn sınıfına) erişim sağlamak için kullanılır. Bu fonksiyon, kalıtım kullanılarak oluşturulan sınıf hiyerarşilerinde, alt sınıflardan üst sınıfların metodlarını ve özelliklerini doğrudan çağırmak için özellikle yararlıdır. Temel kullanım şekli aşağıda verilmiştir.


```python
class TemelSınıf:
    def __init__(self):
        print("TemelSınıf'in inşaçı çağrıldı.")
    def temel_fonksiyon(self):
        print("TemelSınıf'a ait fonksiyon")

class AltSınıf(TemelSınıf):
    def __init__(self):
        super().__init__()  # TemelSınıf'ın __init__ metodunu çağırır
        print("AltSınıf'ın inşaçı çağrıldı.")


# Aşağıda TemelSınıf doğrudan çağrılmamıştır fakat inşacı metodu çalışmıştır.
# Bunun sebebi AltSınıf içinde super() fonksiyonu ile çağrılmasıdır.
print("-------------------------")
ornek_nesne = AltSınıf()
print("-------------------------")
print()

# Aşağıda TemelSınıf' a ait fonksiyon AltSınıf' tan çağrılmıştır.
# super() fonksiyonu AltSınıf tanımlanırken üst sınıftan çağırma işlemini yaptırır.
# super() fonksiyonunun nesneden (ornek_nesne) çağrılması ile ilişkisi yoktur.
print("-------------------------")
ornek_nesne.temel_fonksiyon()
print("-------------------------")
```


super() fonksiyonunun temel kullanımları şunlardır:

- **Üst Sınıfın İnşaçını (Constructor) Çağırmak:** Alt sınıfın --init-- metodu içinde, üst sınıfın --init-- metodunu çağırmak için super() kullanılır.
- **Üst Sınıfın Metodlarını Çağırmak:**  super() kullanarak, alt sınıf metodları içinden üst sınıfın orijinal metodlarına erişilebilir ve çağrılabilir. Bu, metodun genişletilmesi veya özelleştirilmesi gerektiğinde, üst sınıfın implementasyonunun korunmasını sağlar.

## Overriding

Overriding (metod ezme) alt sınıfın,  üst sınıfa ait bir metodu yeniden tanımlaması işlemidir. Bu işlem, alt sınıfın, miras aldığı metodun yerine kendi davranışını koyarak, o metodu kendi ihtiyaçlarına göre özelleştirmesine olanak tanır. Fonksiyonu tamamen değiştirmek de mümkündür fkat buradaki temel yaklaşım üst sınıfa ait fonskiyona yeni özellikler eklemektir.

```python
class TemelSınıf:
    def __init__(self):
        pass       
    def bastan_yazilacak_fonk(self):
        print("TemelSınıf'a ait bastan_yazilacak_fon")
        
    def genisletilecek_fonk(self):
        print(" TemelSınıf'a ait genisletilecek_fonk çağrıldı")

class AltSınıf(TemelSınıf):
    def __init__(self):
        pass
    
    def bastan_yazilacak_fonk(self):
        print("AltlSınıf'a ait fonksiyon baştan yazıldı ")
        
    def genisletilecek_fonk(self):
        super().genisletilecek_fonk()
        print("     TemelSınıf.genisletilecek_fonk' a özellik eklendi")
        
ornek_nesne = AltSınıf()

print("-------------------------")
ornek_nesne.bastan_yazilacak_fonk()
print("-------------------------")
print()
print("-------------------------")
ornek_nesne.genisletilecek_fonk()
print("-------------------------")
```

Yukarıdaki örnekte *bastan_yazilacak_fonk()* AltSınıf' ta *super()* kullanılmadan yeniden yazılmıştır. Bu yapılabilir fakat *genisletilecek_fonk()* yaklaşımı daha doğrudur. Üst sınıftan temel özellikleri alır, üstüne kendi özelliklerini ekler.

Aşağıdaki örnekte tüm araç tiplerinin marka, model ve fiyat özellikleri vardır. Ek olarak kendi içlerinde farklı özellikler alırlar. super() fonksiyonu burada ortak olanları çağırdıktan sonra ek özellikleri yazarken kullanılır.

```python
class Arac:
    def __init__(self, marka, model, fiyat):
        self.marka = marka
        self.model = model
        self.fiyat = fiyat

    def bilgi_goster(self):
        print("--------------")
        print(f"Aracın markası: {self.marka}")
        print(f"Aracın modeli : {self.model}")
        print(f"Aracın fiyatı : {self.fiyat}")
        print("--------------")

class BinekArac(Arac):
    def __init__(self, marka, model, fiyat):
        super().__init__(marka, model, fiyat)
 
        
    def bilgi_goster(self):
        
        print("******************************")     # Başka dillerde çoğunlukla super() öncesi kod yazılmaz.
        print("Binek Araç")                         # Python burada esneklik sağlar.
        super().bilgi_goster()
        print(f"Aracın teker sayısı : 4")
        print(f"Aracın koltuk sayısı : 5")
        print(f"Aracın kapı sayısı : 4")
        print()

class AraziAraci(Arac):
    def __init__(self, marka, model, fiyat):
        super().__init__(marka, model, fiyat)

    def bilgi_goster(self):
        print("******************************")
        print("Arazi Aracı")
        super().bilgi_goster()
        print(f"Aracın teker sayısı : 4")
        print(f"Aracın koltuk sayısı : 5")
        print(f"Aracın kapı sayısı : 5")
        print(f"Aracımız 4 çekerdir.")
        print()
        
class Motorsiklet(Arac):
    def __init__(self, marka, model, fiyat):
        super().__init__(marka, model, fiyat)

    def bilgi_goster(self):
        print("******************************")
        print("Motorsiklet")
        super().bilgi_goster()
        print(f"Aracın teker sayısı : 2")
        print(f"Aracın koltuk sayısı : 1")
        print()
        
        
binek = BinekArac("Toyota", "Corolla", 12000)
arazi = AraziAraci("Suzuki", "Grand Vitara", 10000)
motor = Motorsiklet("Honda", "Transalp", 8000)


binek.bilgi_goster()
arazi.bilgi_goster()
motor.bilgi_goster()
```

## Çoklu kalıtım
Bir programlama dilinde bir sınıfın birden fazla üst sınıftan miras alabilmesi özelliğidir. Bu, bir alt sınıfın birden fazla üst sınıfın özelliklerini ve metodlarını birleştirebilmesine olanak tanır. Çoklu kalıtım, daha karmaşık ilişkileri ve davranışları modellemek için güçlü bir araç olabilir, ancak dikkatli kullanılmadığında karmaşıklığa ve beklenmeyen davranışlara yol açabilir.

Python, çoklu kalıtımı destekleyen dillerden biridir. Bir Python sınıfı, birden fazla üst sınıfı virgülle ayırarak parantez içinde belirterek çoklu kalıtım yapabilir.

```python
class SinifA():
    def ozellik_A(self):
        print("ozellik_A()")
    
    
class SinifB():
    def ozellik_B(self):
        print("ozellik_B()")
        

class AltSinif(SinifA,SinifB):
    def ozellik_alt(self):
        print("ozellik_alt()")
        

ornek_alt = AltSinif()

ornek_alt.ozellik_alt()
```

## Elmas Problemi


Elmas problemi, özellikle nesne yönelimli programlamada, çoklu kalıtımın kullanıldığı durumlarda ortaya çıkan bir yapısal sorundur. Bu problem, bir sınıfın iki farklı sınıftan türemesi ve bu iki sınıfın da aynı üst sınıftan türemiş olması durumunda meydana gelir. Bu yapı, bir elmas şeklini andırdığı için bu ismi alır. Problem, türetilmiş sınıfın, ortak üst sınıftan miras aldığı özelliklere veya metodlara birden fazla yol üzerinden erişebilmesi ve bu durumun belirsizliklere yol açmasıdır.

Bu problemin temelinde yatan sorun, türetilmiş sınıfın hangi yoldan gelen özelliği veya metodu kullanacağının belirsiz olmasıdır. Bu, özellikle metodların override edildiği durumlarda karmaşık hale gelebilir. Çünkü her iki alt sınıf da üst sınıfın metodunu farklı şekillerde override edebilir ve türetilmiş sınıf bu metodları çağırdığında hangisinin kullanılacağı konusunda bir çakışma oluşabilir.

Çoklu kalıtımı destekleyen diller bu problemi çeşitli yollarla çözmeye çalışır. Örneğin, C++ sanal kalıtımı kullanarak, Java ise arayüzler ve sınıf hiyerarşisini düzleştirerek bu tür çakışmaları önlemeye çalışır. Python'da ise Method Resolution Order (MRO) kullanılarak, sınıflar arasındaki hiyerarşiye göre bir çağrı zinciri oluşturulur ve bu zincir, metodların hangi sırayla aranacağını belirler.

## MRO (Method Resolution Order)
Python'da çoklu kalıtım kullanıldığında sınıfların metodlarının nasıl aranıp çağrılacağını belirleyen kurallar dizisidir. Python'da bir sınıfın metodu çağrıldığında, Python bu metodu bulmak için bir arama sırası izler. Bu sıra, o sınıfın kendisinden başlar, ardından miras aldığı sınıfları takip eder. MRO, bu arama sırasını belirler ve böylece hangi metotun çağrılacağını netleştirir.


```python
Asistan.__mro__ # Kalıtım bilgilerini verir.
```

```python
class Personel():
    def __init__(self, maas, yil) -> None:
        self.maas = maas
        self.yil = yil
        
    def gozetmenlik(self):
        print("Gözetmenlik yapıldı")
        
    def bilgi_yazdir(self):
        print("Personel bilgisi")

class Ogrenci():
    def __init__(self, alinan_ders) -> None:
        self.alinan_ders = alinan_ders

        
    def sinava_gir(self):
        print("Sınava girildi")
    
    def bilgi_yazdir(self):
        print("Öğrenci bilgisi")


class Asistan(Personel, Ogrenci):
    def __init__(self, alinan_ders, maas, yil) -> None:
        Ogrenci.__init__(self, alinan_ders)
        Personel.__init__(self, yil, maas)
        
    def ofis_saati(self):
        print("Ofis saati oluşturuldu")
        
    def bilgi_yazdir(self): # Elmas probleminin bir çözüm yolu Override yöntemidir.
        print("Asistan bilgisi")
        
        
asistan_1 = Asistan("Python", 5000, 2020)

Asistan.__bases__
```

# Kapsülleme


Kapsülleme (Encapsulation), nesne yönelimli programlamanın temel prensiplerinden biridir ve bir nesnenin verilerini (özellikler) ve bu veriler üzerinde işlem yapan metodları bir araya getirerek, nesnenin detaylarını dış dünyadan gizlemeyi amaçlar. Kapsüllemenin temel faydaları, veri gizleme ve arayüz soyutlamasıdır.


## Gizleme 

**Erişime göre üye türleri**

Public
- Her yerden erişilebilir üyeler(özellik veya metod).
- API, başka nesneler, fonksiyonlar gibi yapılar tarafından kullanılabilir.
- Varsayılan olarak her üye public olarak tanımlıdır.

Privite
- Yalnızca tanımlandığı sınıftan erişilebilir. (Alt sınıflar da dahil.)
- Sınıfın iç detaylarının dışarıdan erişilebilir olmasını istemediğim durumlarda kullanılır.
- Tanımlamak için üye ismi başına "__" eklenir.

Protected
- Dışarıdan erişilmesin fakat alt sınıflar kullanabilsin.
- Tanımlamak için üye ismi başına "_" eklenir.
- Python için teknik olarak publicten farkı yoktur. (Başka dillerde bu fark mevcuttur.)
- Yazılım geliştiriler arası bir teamüldür.
- Dışarıdan erişilebilir fakat yapılmaması gerekir.
- Kullanımda yapılması gereken bir ayrım.

```python
class UstSinif:
    def __init__(self  ) -> None:
        self.k_adi = "Kullanıcı adı"
        self._ad = "Ad-Soyad"
        self.__parola = "Parola" # Dışarıdan erişilemez.
        
    def public_metod(self):
        return "Bu bir public metod"

    def _protected_metod(self):
        return "Bu bir protected metod"

    def __private_metod(self):
        return "Bu bir private metod"
        
class AltSinif(UstSinif):
    def __init__(self):
        super().__init__()
        self.alt_sinif_degiskeni = "Ben AltSınıf' a ait bir değişkenim"
        
    def alt_sinif_metod(self):
        print("AltSinif metodu çağrıldı.")
        
    def erisim(self):
        print(self.ad)
        

def dıs_erisim(nesne):
    print("nesne.k_adi:", nesne.k_adi)
    print("nesne.ad:", nesne.ad)
    print("nesne.parola:", nesne.parola)
        
ust_nesne = UstSinif()
alt_nesne = AltSinif()

def uye_erisim(uye):
    print(uye)


# Aşağıdaki kod çalışır fakat önerilmez.
ust_nesne._ad # Dışarıdan doğrudan erişildi.
uye_erisim(ust_nesne._ad) # Dış fonksiyon erişti.


# Alttaki kod hata verir.
ust_nesne.__parola 
uye_erisim(ust_nesne._ad) 

ust_nesne._ad
alt_nesne.erisim()
dıs_erisim(ust_nesne)
```

```python
# Bu örnekte ses çıkarma tüm hayvanların ortak özelliğidi. 
# Fakat detaylarda yapılan işler değişir. 

class EvcilHayvan:
    def __init__(self, yas) -> None:
        self.yas = yas
        
    def ses_cikar(self):
        print("Evcil havyanın sesi:")
        
class Kus(EvcilHayvan):
    def __init__(self, yas) -> None:
        super().__init__(yas)
        
    def ses_cikar(self):
        super().ses_cikar()
        print("cikcik")
        
class Kedi(EvcilHayvan):
    def __init__(self, yas) -> None:
        super().__init__(yas)
        
    def ses_cikar(self):
        super().ses_cikar()
        print("miyav")
kus = Kus(2)
kedi = Kedi(3)

kedi.ses_cikar()
kus.ses_cikar()
```

```python
class Sekil(object):
    def __init__(self) -> None:
        pass

    def alan_hesapla(self):
        print("Şeklinizin Alanı:")
       
        
class Ucgen(Sekil):
    def __init__(self, a, b, c) -> None:
        super().__init__()
        self.a = a
        self.b = b 
        self.c = c
               
    def alan_hesapla(self):
        super().alan_hesapla()
        u = (self.a+self.b+self.c)/2
        alan = (u*(u-self.a)*(u-self.b)*(u-self.c))**(1/2)
        print(alan)
        return alan
    
    
class Kare(Sekil):
    def __init__(self, a) -> None:
        super().__init__()
        self.a = a
        
    def alan_hesapla(self):
        super().alan_hesapla()
        alan = self.a**2
        print(alan)
        return alan
    
def sekil_alan_hesabla(*arg):
    for sekil in [*arg]:
        sekil.alan_hesapla()
        
        
ucgen = Ucgen(3,4,5)
kare = Kare(5)
ucgen.alan_hesapla()
kare.alan_hesapla()
sekil_alan_hesabla(kare, ucgen)
```

# Soyutlama

Soyutlama (Abstraction), nesne yönelimli programlamada, karmaşık bir sistem içindeki önemli detayları vurgulayıp, önemsiz detayları göz ardı ederek sistem üzerinde çalışmayı kolaylaştıran bir prensiptir. Soyutlama, bir programın veya sistemin nasıl çalıştığını anlamayı ve kullanmayı kolaylaştırırken, iç işleyişinin karmaşık detaylarını gizler.

```python
class Arac:
    def __init__(self, yakit_turu, uretim_yili) -> None:
        self.yakit_turu = yakit_turu
        self.uretim_yili = uretim_yili
        
        
    def hız_hesapla(self, hiz_turu):
        """Bu fonksiyon tanımlı aracın birim zamanda gidilen mesafesini hesaplar.
        """
        pass
    
class KaraTasiti(Arac):
    def __init__(self, yakit_turu, uretim_yili, tekerlek_sayisi, motor_gucu) -> None:
        super().__init__(yakit_turu, uretim_yili)
        
    def hız_hesapla(self, hiz_turu):
        print("Hızımız hesaplandı")
    
    # Araç dizel, benzinli veya elektrikli olabilir. 
    def motoru_calistir(self):
        pass
    
    def motoru_durdur(self):
        pass
        
class BinekArac(KaraTasiti):
    def __init__(self, yakit_turu, uretim_yili, tekerlek_sayisi, motor_gucu, vites_tipi) -> None:
        super().__init__(yakit_turu, uretim_yili, tekerlek_sayisi, motor_gucu) 
        self.vites_tipi = vites_tipi       
    
    
    def motoru_calistir(self):
        if self.vites_tipi == "Elektrikli":
            print("Düğmeye bas.")
        elif self.vites_tipi == "Otomatik":
            print("Anahtarla çalıştır")
    
        
        
    
araba = BinekArac("yakit_turu", "uretim_yili", "tekerlek_sayisi", "motor_gucu", "Elektrikli")


# Soyut bir sınıfta tanımlanmış bir metodun alt sınıflarda kesinlikle kullanılması gerekir.
araba.motoru_durdur() 
    
```

```python
from abc import ABC, abstractmethod

class OdemeYontemi(ABC):
    
    @abstractmethod
    def ode(self):
        print("Soyut sınıf metodu")


class NakitOde(OdemeYontemi):
    def ode(self): # Override
        print("nakit odendi")


class KrediKartiylaOde(OdemeYontemi):
    def yazdir():
        print("aaa")



class soyut_alt_odeme(OdemeYontemi):
    """OdemeYontemi sınıfından türemiş bir AbstractClass

    Args:
        OdemeYontemi (ABC): Üst sınıf
    """
    pass

def odeme(tur):
    tur.ode()



nakit = NakitOde()
kredi = KrediKartiylaOde()

odeme(nakit)
odeme(kredi)
```
