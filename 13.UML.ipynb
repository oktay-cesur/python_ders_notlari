```python
from abc import ABC

class Hayvan(ABC):
    def __init__(self, yas: int, cinsiyet: str) -> None:
        self.yas = yas
        self.cinsiyet = cinsiyet
        
    def memeli_mi(self):
        pass
    
    
class Ordek(Hayvan):
    gaga_rengi = "Sar"
    def __init__(self, yas: int, cinsiyet: str) -> None:
        super().__init__(yas, cinsiyet)
        
    def memeli_mi(self):
        return False
    
    def yuz(self):
        print("rdeimiz yzd")
    
    def vakvak(self):
        print("vakvak")


class Balik(Hayvan):
    def __init__(self, yas: int, cinsiyet: str, boyut: int, yenir_mi: bool) -> None:
        super().__init__(yas, cinsiyet)
        self.boyut = boyut
        self.yenir_mi = yenir_mi
        
    def memeli_mi(self):
        return False
    
    def yuz():
        print("Balmz yzd")

       
class Zebra(Hayvan):
    def __init__(self, yas: int, cinsiyet: str, vahsi_mi: bool) -> None:
        super().__init__(yas, cinsiyet)
        self.vahsi_mi = vahsi_mi
    
    def memeli_mi(self):
        return True
    
    def kos(self):
        print("Zrafa kotu")

```

Araba - Motor - Tekerlek

- Arabann almak iin motora ve 4 tane tekere ihtiyac var
- Motorun hacmi

```python
class Motor:
    def __init__(self, hacim: float) -> None:
        self.hacim = hacim
    
    def calis(self):
        print("Motor alt")
        
class Tekerlek:
    def __init__(self, marka: str) -> None:
        self.marka = marka
        
    def don(self):
        print("Tekerlek dnd")
        
#  Arabann oluabilemesi iin tekerleklere ve motora ihtiyac var.
class Araba:
    def __init__(self, marka: str, hacim: float, tekerlek1: Tekerlek,
                 tekerlek2: Tekerlek,tekerlek3: Tekerlek,tekerlek4: Tekerlek) -> None:
        self.marka = marka
        # Motor araba ile beraber tanmland.
        self.motor = Motor(hacim)  
        # Tekerlekler arabadan nce tanmland.
        self.tekerlek1 = tekerlek1
        self.tekerlek2 = tekerlek2
        self.tekerlek3 = tekerlek3
        self.tekerlek4 = tekerlek4
        
    def git(self):
        self.motor.calis()
        self.tekerlek1.don()
        self.tekerlek2.don()
        self.tekerlek3.don()
        self.tekerlek4.don()
        

t1 = Tekerlek("Lassa")
t2 = Tekerlek("Lassa")
t3 = Tekerlek("Lassa")
t4 = Tekerlek("Lassa")

araba = Araba("Opel", 2.5, t1, t2, t3, t4)
```

```python
araba.git()
```

```python
class Mutfak:
    def __init__(self, m2: int) -> None:
        self.m2 = m2
    
    def yemek_hazrla(self):
        print("Yemek hazrland")
        
        
class Banyo:
    def __init__(self, m2: int) -> None:
        self.m2 = m2
    
    def dus_al(self):
        print("Banyoya girildi")   
    
    
class Oda:
    def __init__(self,ad, m2: int) -> None:
        self.ad = ad
        self.m2 = m2
    
    def otur(self):
        print(f"inde bulunulan oda: {self.ad}")  
        

class PostaKutusu:
    def __init__(self, sokak_adi) -> None:
        self.sokak_adi = sokak_adi
    
    def posta_al(self):
        print("Posta alnd")
        

class Kredi:
    def __init__(self, miktar: int, taksit_sayisi: int) -> None:
        self.miktar = miktar
        self.taksit_sayisi = taksit_sayisi
        self.kalan_taksit = taksit_sayisi
        self.ev = None
    
    def ev_tanimla(self, ev: object):
        self.ev = ev
        
    def kredi_ode(self):
        if self.kalan_taksit > 0:
            print("kredinizin 1 taksidi dendi")
            self.kalan_taksit -= 1
        else:
            print("Taksitler bitti")
            
            
   
   
# Evimizin odalara ihtiyac var. - composition
# Odalar ev ile beraber tanmlanmal. 
# Posta kutusu evin dnda tanmlanmal.
# Evi tanmlamak iin posta kutusuna da ihtiya var. - aggregetion   
# Kredi ile ev balantl ama biri dierinin paras deil.
            
class Ev:
    def __init__(self, oda_adi, oda_m2, banyo_m2, mutfak_m2, posta_kutusu: PostaKutusu ) -> None:
        self.oda = Oda(oda_adi, oda_m2)
        self.banyo = Banyo(banyo_m2)
        self.mutfak = Mutfak(banyo_m2)
        self.posta_kutusu = posta_kutusu
        self.kredi = None
        
    def kredi_cek(self, kredi: Kredi):
        self.kredi = kredi
        self.kredi.ev_tanimla(self)
       

posta_kutusu = PostaKutusu("Vatan Caddesi")        


ev1 = Ev("Salon", 20, 10, 15, posta_kutusu)
kredi1 = Kredi(1000000, 120)


class kredi_olustur:
    def __init__(self,ev: Ev, kredi: Kredi) -> None:
        ev.kredi_cek(kredi)
        kredi.ev_tanimla(ev)

```

- Ktphane - Kitap - Yazar - Konu


```python
class Kutuphane:
    def __init__(self, ad) -> None:
        self.kutuphane_adi = ad
        self.kitaplar = []
        self.yazarlar = []
        self.konular = []
        
    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)
        
    def yazar_ekle(self, yazar):
        self.yazar.append(yazar)  
        
    def konu_ekle(self, konu):
        self.konu.append(konu)  
        
    def imza_gunu(self, yazar, kitap):
        if kitap in yazar.kitaplar:
            print("mza gnmze ho geldiniz")
            
            
class Yazar:
    def __init__(self, yazar_adi) -> None:
        self.yazar_adi = yazar_adi
        self.kitaplar = []
        
    def kitap_ekle(self, kitap):
        if kitap not in self.kitaplar:
            self.kitaplar.append(kitap)
        else:
            print("Kitap zaten mevcut.")
            
    def imza_gunu(self, kutuphane, kitap):
        if kitap in self.kitaplar:
            print("mza gnmze ho geldiniz")   

         
class Kitap:
    def __init__(self, ad, kutuphane, yazarlar, konular) -> None:
        self.kitap_adi = ad
        self.konular = konular
        self.yazarlar = yazarlar
        self.adet = 0
        self.kutuphane = kutuphane
        
    def kitap_oku(self):
        print("Kitap okundu")
        
    def yazar_ekle(self, yazar):
        self.yazar.append(yazar) 

    def konu_ekle(self, konu):
        self.konu.append(konu)

        
class Konu:
    def __init__(self, konu_adi) -> None:
        self.konu_adi = konu_adi
        self.kitaplar = []
        
    def kitap_ekle(self, kitap):
        if kitap not in self.kitaplar:
            self.kitaplar.append(kitap)
        else:
            print("Kitap zaten mevcut.")


def kitap_olustur(kitap_adi, yazarlar: list, konular: list):
    kitap = Kitap(kitap_adi, None, yazarlar, konular)
    
    for yazar in yazarlar:
        yazar.kitaplar.append(kitap)
        
    for konu in konular:
        konu.kitap_ekle(kitap)
        
def kutuphaneye_kitap_ekle(kitap: Kitap, kutuphane: Kutuphane, adet: int):
    kitap.kutuphane = kutuphane
    kitap.adet += adet
    kutuphane.kitap_ekle(Kitap)

    for konu in kitap.konular:
        if konu not in kutuphane.konular:
            kutuphane.konu_ekle(konu)

    for yazar in kitap.yazarlar:
        if yazar not in kutuphane.yazarlar:
            kutuphane.yazar_ekle(yazar)
```

- niversite - Faklte - Blm (opsiyonel - Akademisyen - renci - Ders)


```python
class Bolum:
    def __init__(self, ad):
        self.ad = ad
        self.akademisyenler = []

    def bilgi_ver(self):
        return f"Blm: {self.ad}"


class Fakulte:
    def __init__(self, ad):
        self.ad = ad
        self.bolumler = []  # Faklte snf, blmler listesine sahiptir.

    def bolum_ekle(self, bolum_adi):
        bolum = Bolum(bolum_adi)
        self.bolumler.append(bolum)

    def bilgi_ver(self):
        bolum_listesi = [bolum.bilgi_ver() for bolum in self.bolumler]
        return f"{self.ad} adl fakltede u blmler bulunmaktadr: {', '.join(bolum_listesi)}"


class Universite:
    def __init__(self, ad):
        self.ad = ad
        self.fakulteler = []  # niversite snf, faklteler listesine sahiptir.

    def fakulte_ekle(self, fakulte_adi):
        fakulte = Fakulte(fakulte_adi)
        self.fakulteler.append(fakulte)

    def bilgi_ver(self):
        fakulte_listesi = [fakulte.bilgi_ver() for fakulte in self.fakulteler]
        return f"{self.ad} adl niversitede u faklteler bulunmaktadr: {', '.join(fakulte_listesi)}"


class Akademisyen:
    def __init__(self, ad, uzmanlik) -> None:
        self.akademisyen_adi = ad
        self.uzmanlik = uzmanlik
        self.bolum: Bolum
        


  
def bolume_akademisyen_ata(bolum: Bolum, akademisyen: Akademisyen):
    bolum.akademisyenler.append(akademisyen)
    akademisyen.bolum = bolum    
    



universite = Universite("rnek niversite")
universite.fakulte_ekle("Mhendislik Fakltesi")
universite.fakulteler[0]
universite.fakulteler[0].bolum_ekle("Bilgisayar Mhendislii")
universite.fakulteler[0].bolum_ekle("Elektrik Mhendislii")
universite.fakulteler[0].bolum_ekle("naat Mhendislii")

# niversite nesnesi oluturuluyor ve faklteler ekleniyor


print(universite.bilgi_ver())


akademisyen1 = Akademisyen("Cem Say", "Doal Dil leme")

bolume_akademisyen_ata(universite.fakulteler[0].bolumler[0], akademisyen1)


```
